```python


def get_table_bounding_boxes(page):
    bounding_boxes = []
    min_x = 10000 
    min_y = 10000
    max_x = -10000
    max_y = -10000
    for n in range(len(page['regions'])):
        table = page['regions'][n]
        polygon = table['polygon']
        # Calculate min and max x and y coordinates for the current polygon
        min_x = min(min_x, np.min(polygon[:, 0]))
        min_y = min(min_y, np.min(polygon[:, 1]))
        max_x = max(max_x, np.max(polygon[:, 0]))
        max_y = max(max_y, np.max(polygon[:, 1]))
        
        # Append the bounding box as a tuple
#        bounding_boxes.append((min_x, min_y, max_x, max_y))
    
    return (min_x, min_y, max_x, max_y)

```

```python
def plot_page_with_regions(image_path, page):
    # Read the image
    thickness = 5
    point_radius = 10
    cell_colour = (0, 0, 255)
    image = cv2.imread(image_path)
    
    if image is None:
        raise ValueError(f"Image not found at the path: {image_path}")
    
    # Convert vertices to a numpy array
#    print(len(page['regions']),page['regions'])
    for reg in page['regions']:
        if reg['regiontype'] == "TableRegion":
            region_colour = (0, 0, 0)
        elif reg['regiontype'] == "TextRegion":
            region_colour = (0, 255, 0)
        
        for cell in reg['table']:
            xmax, ymax = cell['polygon'].max(axis=0)
            xmin, ymin = cell['polygon'].min(axis=0)
            cell_left_top = (xmin, ymin)
            cell_left_bottom = (xmin, ymax)
            cell_right_top = (xmax, ymin)
            cell_right_bottom = (xmax, ymax)
            cv2.circle(image, cell_left_top, radius=point_radius, color=cell_colour)            
            cv2.circle(image, cell_left_bottom, radius=point_radius, color=cell_colour)            
            cv2.circle(image, cell_right_top, radius=point_radius, color=cell_colour)            
            cv2.circle(image, cell_right_bottom, radius=point_radius, color=cell_colour)            
        pts = reg['polygon']
        cv2.polylines(image, [pts], isClosed=True, color=region_colour, thickness=thickness)
            
    # Convert BGR to RGB for displaying with matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    
    # Plot the image
    plt.figure(figsize=(10, 10))
    plt.imshow(image_rgb)
    plt.axis('off')  # Hide axis
    plt.show()
```

```python
def write_yolo_entry(file, category, seg, width, height):
    seg = seg.flatten().tolist()
    result = [min(num / width, 1.0) if i % 2 == 0 else min(num /height, 1.0)  for i, num in enumerate(seg)]
    reg_coords = ""
    for coord in result:
        reg_coords = reg_coords + " {}".format(str(coord))
    file.write(f"{category} {reg_coords}\n")   
```

```python
def show_yolo_entry(category, seg, width, height):
    seg = seg.flatten().tolist()
    result = [min(num / width, 1.0) if i % 2 == 0 else min(num /height, 1.0)  for i, num in enumerate(seg)]
    reg_coords = ""
    for coord in result:
        reg_coords = reg_coords + " {}".format(str(coord))
    print(f"{category} {reg_coords}")
    print(f"{seg}")
```

```python

def createYOLOsegmentCells(dataset, yolo_path, selection, imgsz=(640,640), resize=True):
    if not os.path.exists(yolo_path):
        os.makedirs(yolo_path)
    if not os.path.exists(os.path.join(yolo_path, "images")):
        os.makedirs(os.path.join(yolo_path, "images"))
    if not os.path.exists(os.path.join(yolo_path, "labels")):
        os.makedirs(os.path.join(yolo_path, "labels"))
    if resize:
        for idx,id in enumerate(selection):
            source = dataset.__getfullname__(id)
            destination = f"{yolo_path}/images/img{idx}.jpg"
            img = cv2.imread(source)
            resized_img = cv2.resize(img, imgsz)
            cv2.imwrite(destination, resized_img)
    for idx ,id in enumerate(selection):
        name, size = dataset.__getsize__(id)
        image_width = size[1]
        image_height = size[0]
        page = dataset.__getXMLitem__(id)
        if len(page['regions']) > 0:
            file_object = open(f"{yolo_path}/labels/img{idx}.txt", "w")
            regions = []
            for i, reg in enumerate(page['regions']):
                n_reg = {}
                n_reg['index'] = i
                xmax, ymax = reg['polygon'].max(axis=0)
                xmin, ymin = reg['polygon'].min(axis=0)
                n_reg['topcorner'] = (xmin, ymin)
                regions.append(n_reg)
            s_regions = sorted(regions, key=lambda x: (x['topcorner'][1], x['topcorner'][0]), reverse=False)
            if len(page['regions']) == 1:
                if 'table' in page['regions'][s_regions[0]['index']]:
                    for cell in page['regions'][s_regions[0]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)
            elif len(page['regions']) == 2:
                if 'table' in page['regions'][s_regions[0]['index']]:
                    for cell in page['regions'][s_regions[0]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)
                if 'table' in page['regions'][s_regions[1]['index']]:
                    for cell in page['regions'][s_regions[1]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)
            elif len(page['regions']) == 3:
                write_yolo_entry(file_object, annotations['title'], page['regions'][s_regions[0]['index']]['polygon'], image_width, image_height)
                write_yolo_entry(file_object, annotations['header'], page['regions'][s_regions[1]['index']]['polygon'], image_width, image_height)
                if 'table' in page['regions'][s_regions[2]['index']]:
                    for cell in page['regions'][s_regions[2]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)
            elif len(page['regions']) == 4:
                write_yolo_entry(file_object, annotations['header'], page['regions'][s_regions[0]['index']]['polygon'], image_width, image_height)
                write_yolo_entry(file_object, annotations['header'], page['regions'][s_regions[1]['index']]['polygon'], image_width, image_height)
                
                if 'table' in page['regions'][s_regions[2]['index']]:
                    for cell in page['regions'][s_regions[2]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)

                if 'table' in page['regions'][s_regions[3]['index']]:
                    for cell in page['regions'][s_regions[3]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)
            elif len(page['regions']) == 5:
                write_yolo_entry(file_object, annotations['title'], page['regions'][s_regions[0]['index']]['polygon'], image_width, image_height)
                write_yolo_entry(file_object, annotations['title'], page['regions'][s_regions[1]['index']]['polygon'], image_width, image_height)
                if 'table' in page['regions'][s_regions[2]['index']]:
                    for cell in page['regions'][s_regions[2]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)

                write_yolo_entry(file_object, annotations['header'], page['regions'][s_regions[3]['index']]['polygon'], image_width, image_height)
                if 'table' in page['regions'][s_regions[4]['index']]:
                    for cell in page['regions'][s_regions[4]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)

            elif len(page['regions']) == 6:
                write_yolo_entry(file_object, annotations['title'], page['regions'][s_regions[0]['index']]['polygon'], image_width, image_height)
                write_yolo_entry(file_object, annotations['title'], page['regions'][s_regions[1]['index']]['polygon'], image_width, image_height)
                write_yolo_entry(file_object, annotations['header'], page['regions'][s_regions[2]['index']]['polygon'], image_width, image_height)
                write_yolo_entry(file_object, annotations['header'], page['regions'][s_regions[3]['index']]['polygon'], image_width, image_height)
                if 'table' in page['regions'][s_regions[4]['index']]:
                    for cell in page['regions'][s_regions[4]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)

                if 'table' in page['regions'][s_regions[5]['index']]:
                    for cell in page['regions'][s_regions[5]['index']]['table']:
                        write_yolo_entry(file_object, annotations['cell'], cell['polygon'], image_width, image_height)
            else:
                print("Annotation error", idx, dataset.__getfullname__(idx), "regions:", len(page['regions']) )
                
            file_object.close()
```
